```python
import numpy as np,pandas as pd
from sklearn import datasets
from keras.models import Sequential
from keras.layers import Dense, Dropout,Input
from sklearn.model_selection import train_test_split
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn import metrics
```

```python
data_iris = datasets.load_iris()
data = pd.DataFrame(data_iris['data'])
label = pd.DataFrame((OneHotEncoder().fit_transform(data_iris['target'].reshape(-1, 1))).toarray())
X_train,X_val,Y_train,Y_val = train_test_split(data,label, test_size = 0.3, random_state = 0)
del data, label
```

```python
def NN_model():
#     model = K.initializers.glorot_uniform(seed=4)
    model = Sequential()
    model.add(Dense(40, input_dim=4,activation='relu'))
#     model.add(Dropout(0.15))
    model.add(Dense(20, activation='relu'))
#     model.add(Dropout(0.5))
#     model.add(Dense(20, kernel_initializer='uniform', activation='relu'))
#     model.add(Dropout(0.5))
    model.add(Dense(3,activation='softmax'))

    model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
    return model

nn_model = KerasClassifier(build_fn=NN_model, verbose=1, epochs=200); 
nn_model.fit(X_train,Y_train.values)

out_of_fold_nn = nn_model.predict(X_train)
predictions_nn = nn_model.predict(X_val)
```

```python
YY_val = []
```

```python
for idx1 in np.array(Y_val):
    for idx2 in range(3):
        if idx1[idx2] == 1:
            YY_val.append(idx2)
```

```python
metrics.f1_score(YY_val, predictions_nn,average='macro')
```
